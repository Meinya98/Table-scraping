Table scraping exercise<br>
Use "List of largest companies in Indonesia" from Wikipedia

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_Indonesia'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[0] #Use find_all instead of find because there are multiple tables
```

Table 1 out of 2 (as of 2025/01/14)

```python
table1 = soup.find_all('table')[0] #define the table1 from the first table
print(table1)
```

```python
#extract the title from the table1
title1 = table1.find_all('th') #'th' because title is inside the 'th' tag
print(title1)
```

```python
#clean the title
table_title1 = [title.text.strip() for title in title1] #use strip directly inside the loop
print(table_title1)
```

```python
#create dataframe using pandas
df_table1 = pd.DataFrame(columns=table_title1)
df_table1
```

```python
column_data1 = table1.find_all('tr') #'tr' is the tag for each row
column_data1
```

```python
for row in column_data1[1:]:
    row_data1 = row.find_all('td') #'td' is the tag for each data inside a row
    individual_row_data1 = [data.text.strip() for data in row_data1]
    #print(individual_row_data1) #from printing, the first row is empty, thus we need to start from 1 
    length = len(df_table1) #look at how many rows in the dataframe
    #print(length)
    df_table1.loc[length] = individual_row_data1 #put the data in the next position
    #print(individual_row_data1)
```

```python
df_table1
```

```python
df_table1.to_csv(r'D:\Informatics Exercises\Table-scraping-from-web\Table-scraping\TopIndoCompanies.csv', index = False)
```
